```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
import seaborn as sns
from sklearn import cluster
%matplotlib inline
```

```python
df=pd.read_csv("c_3.csv")
le = preprocessing.OneHotEncoder()
le.fit(df.loc[:,"BenefitName"].to_numpy().reshape(-1,1))
x=pd.read_csv("dataset_new.csv")
x=x.set_index("PlanId")
x=(x!=0)*1
ben_list=le.inverse_transform(np.diag(np.ones(37))).flatten().tolist()
x = x.set_axis(ben_list, axis=1, inplace=False)
x=x.reset_index()
```

```python
def jaccard(id,df):
    temp=[]
    temp2=np.array(df.iloc[:,1:38])
    row=np.array(x.iloc[:,1:38][x.loc[:,"PlanId"]==id])
    n=temp2.shape[0]
    for i in range(n):
        intersection=np.sum(row*temp2[i,:])
        union=np.sum(row+temp2[i,:])-intersection
        temp.append(1-(intersection/union))
    return temp
```

```python
ID="11512NC0060019"
state="AL"
age=24
df2=df[df.loc[:,"StateCode"]==state]
df2=df2[df2.loc[:,"Age"]==age]
df2=le.transform(df2.loc[:,"BenefitName"].to_numpy().reshape(-1,1))
df2=pd.DataFrame.sparse.from_spmatrix(df2)
df2.loc[:,"PlanId"]=df.loc[:,"PlanId"]
df2=df2.groupby(by="PlanId").sum()
df2=(df2!=0)*1
ben_list=le.inverse_transform(np.diag(np.ones(37))).flatten().tolist()
df2 = df2.set_axis(ben_list, axis=1, inplace=False)
df2=df2.reset_index()
df2["Jaccard"]=jaccard(ID,df2)
```
